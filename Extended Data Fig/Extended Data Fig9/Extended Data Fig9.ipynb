```python
import pandas as pd
import os,sys,glob,random
from tqdm import tqdm
sys.path.append(r"D:\jiaozhuolei\python\code\neuron-vis\neuronVis")  # the neuron-vis folder
import Visual as nv
```

```python
# load the neuron infor
dfneuron = pd.read_excel("compare_mouselight_data.xlsx",index_col=0)
```

```python
dfneuron
```

```python
# load the view angle infor
left = dict(eye=(0.0,0.0,-15000),center=(0,0,0),up=(0,-1,0))
views = [left]
labels = ["left"]
```

```python
dfneuron.loc[dfneuron.cluster ==1]["distance"].tolist()[0]
```

```python
dfneuron
```

```python
Datapth = ".\\swc"
neuronvis = nv.neuronVis(size=(1600,1000),renderModel=1)
neuronvis.render.setBackgroundColor([1,1,1])
for cluster in tqdm(dfneuron.cluster.to_list()):
    neuron1 = dfneuron.loc[dfneuron.cluster ==cluster]["neuron1"].tolist()[0]
    file1 = os.path.join(Datapth,neuron1+".swc")
    neuronvis.addNeuron(file1,somaColor = [0.0,0.0,0.0],axonColor= [1.0,0.0,0.0],dendriteColor=[1.0,0.0,0.0])

    neuron2 = dfneuron.loc[dfneuron.cluster ==cluster]["neuron2"].tolist()[0]
    file2 = os.path.join(Datapth,neuron2+".swc")
    neuronvis.addNeuron(file2,somaColor = [0.0,0.0,0.0],axonColor= [0,0,1],dendriteColor=[0,0,1])    
    neuronvis.render.setBackgroundColor([1,1,1])
    neuronvis.render.setLineWidth(2)

    distance = dfneuron.loc[dfneuron.cluster ==cluster]["distance"].tolist()[0]
    
    for label,view in zip(labels,views):
        neuronvis.render.setLookAt(eye=view["eye"],center=view["center"],up=view["up"]) 
        neuronvis.render.savepng("Subtype_%d_%s_%s_%s_%s.png"%(cluster,neuron1,neuron2,distance,label))
    neuronvis.clear(neurons=True,regions=True)
    neuronvis.neurons.clear()
neuronvis.render.closeWindow()
```
