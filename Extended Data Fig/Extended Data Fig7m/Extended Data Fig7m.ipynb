```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfplot = pd.read_excel("bilateral_neuron_enrich.xlsx",index_col=0)
dfcorr = dfplot.corr()
```

```python
sns.clustermap(dfcorr,figsize=(8,8),cmap = "YlGnBu",vmax = 1,linewidths=2, linecolor='white')
plt.savefig("Extended Data Fig7m.jpg",dpi = 300)
plt.savefig("Extended Data Fig7m.pdf",dpi = 300)
```
