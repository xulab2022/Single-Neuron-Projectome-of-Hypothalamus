```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
plt.rc("font",family="Arial")
```

```python
df = pd.read_excel("Orexin_chose.xlsx",index_col=0)
```

```python
arealist = ["RSP","MO","SS","GU","VISC","AUD","VIS","ACA","PL","ILA","ORB","AI","FRP","PTLp","TEa","PERI","ECT","MOB","AOB","AON","TT","DP","PIR","NLOT","COA","PAA","TR"]
```

```python
dfdata =df[["AI","PIR","AON","TT","MOB","ACA","VIS","RSP","ENT","ORB","SS","MO"]]
```

```python
dfcorr = dfdata.corr()
sns.set()

g=sns.clustermap(dfcorr, fmt="d",linewidths=1,cmap='YlGnBu',figsize=(6,6))
ax = g.ax_heatmap
plt.savefig("corr_orexin_neruons_iso.jpg",dpi =600,format="jpg")
plt.savefig("corr_orexin_neruons_iso.pdf",dpi =600,format="pdf")
```
