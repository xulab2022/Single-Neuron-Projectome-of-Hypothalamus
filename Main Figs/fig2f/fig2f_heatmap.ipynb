```python
import pandas as pd
import numpy as np
import os,glob
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42
plt.rc("font",family="Arial")
```

```python
dfplot = pd.read_csv("plot.csv",index_col=0)
```

```python
dfcolor = pd.read_csv("plot_color.csv",index_col=0)
```

```python
colorlist = dfcolor.loc[:,"color"].tolist()
```

```python
g = sns.clustermap(dfplot,col_cluster=False,row_cluster=True,col_colors=colorlist,figsize=(16,5),cmap = "bwr",xticklabels=False,vmax = 0.1)

ax_heatmap = g.ax_heatmap

for label in ax_heatmap.get_yticklabels():
    label.set_fontsize(20)
    label.set_rotation(0) 

pos = g.ax_col_colors.get_position()

# 设定新的位置，增加高度（第四个参数 height）
g.ax_col_colors.set_position([pos.x0, pos.y0, pos.width, pos.height * 4])
# plt.tight_layout()
plt.savefig("fig2f.pdf",dpi = 300)
```
