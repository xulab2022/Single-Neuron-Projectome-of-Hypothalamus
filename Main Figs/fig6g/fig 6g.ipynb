```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("Agrp_Pomc_ARH_cluster_4.xlsx",index_col=0)
```

```python
dfplot = df.pivot_table(index="geno",columns="cluster",aggfunc="count",values="neurons")
```

```python
fig, ax = plt.subplots(figsize = (4,2.4))
sns.set(font_scale=1.9)
sns.heatmap(data = dfplot,cmap="YlGnBu",vmin=0,
                 vmax=50,linewidths=1, linecolor='white',annot=True,fmt="g")
plt.tight_layout()
# plt.savefig("heatmap_for_intra_three_gene.jpg",dpi =300)
plt.savefig("heatmap_for_fig6g.pdf",dpi =300,format="pdf")
plt.savefig("heatmap_for_fig6g.jpg",dpi =300,format="jpg")
```
